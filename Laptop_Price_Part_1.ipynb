```python
#categorize processor types based on the input text
def fetch_processor(text):
  if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
    return text
  elif text.split()[0] == 'Intel':
    return 'Other Intel Processor'
  else:
    if text[4] == 'E':
      return 'AMD E Series Processor'
    elif text[4] == 'A':
      return 'AMD A Series Processor'
    elif text[4] == 'D':
      return 'AMD D Series Processor'
    elif text[4] == 'F':
      return 'AMD F Series Processor'
    else:
      return 'AMD Ryzen Series'
```

# Reduce Categories:
Group rare manufacturers in Company into an "Other" category.

Simplify OpSys by grouping similar operating systems (e.g., Windows, macOS, Linux, Others).

Use lambda functions for grouping categories.


```python
#Group rare manufacturers in Company into an "Other" category.
df['Company'].value_counts()
```

```python
#  Group rare manufacturers in Company into an "Other" category.

def group_rare_manufacturers(df, column_name='Company', threshold=0.03):


    # Calculate manufacturer proportions
    manufacturer_pro = df['Company'].value_counts(normalize=True)

    # Identify manufacturers below the threshold
    rare_manufacturers = manufacturer_pro[manufacturer_pro < threshold].index

    # Replace rare manufacturers with "Other"
    df['Company'] = df['Company'].replace(rare_manufacturers, 'Other')
    return df

# Example usage (replace 'Company' if your column has a different name):
df = group_rare_manufacturers(df)
```

```python
df['Company'].value_counts()
```

```python
#Simplify OpSys by grouping similar operating systems (e.g., Windows, macOS, Linux, Others).
df['OpSys'].value_counts()
```

```python
#  Simplify OpSys by grouping similar operating systems (e.g., Windows, macOS, Linux, Others).

def simplify_os(os_name):
    if 'Windows' in os_name:
        return 'Windows'
    elif 'macOS' in os_name:
        return 'macOS'
    elif 'Linux' in os_name:
        return 'Linux'
    else:
        return 'Other'

df['OpSys'] = df['OpSys'].apply(simplify_os)
df['OpSys'].value_counts()
```
